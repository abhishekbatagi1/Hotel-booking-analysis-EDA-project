```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
df = pd.read_csv('Hotel Bookings.csv')
```

```python
pd.set_option('display.max_columns', None)  # Set to None to display all columns
df.sample(5)
```

```python
plt.figure(figsize = (10,5))

missing_values = df.isnull()
sns.heatmap(missing_values, cmap='inferno')

plt.title("Missing Data Heatmap")
plt.show()
```

```python

df=df.fillna(df['agent'].mean(),axis=1)
```

```python
df=df.drop(columns='company', axis=1)
```

```python
df= df.fillna(value=df['agent'].mean(),axis=1)
```

```python
df
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df['children'].fillna(df['children'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df[['children', 'agent']] = df[['children',  'agent']].astype('int64')
```

```python
pd.set_option('display.max_columns', None)  # Set to None to display all columns
df.sample(5)
```

```python
df['total_people'] = df['adults']+df['children']+df['babies']
df['total_stay'] = df['stays_in_weekend_nights']+df['stays_in_week_nights']
```

```python
df.columns
```

```python
df.groupby('hotel')['arrival_date_month'].count()
```

```python
month_mapping = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
                 'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}

# Convert 'arrival_date_month' to numeric representation
df['arrival_date_month'] = df['arrival_date_month'].map(month_mapping)
```
